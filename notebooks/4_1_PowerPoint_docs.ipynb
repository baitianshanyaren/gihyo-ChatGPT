```python
from pptx import Presentation
from pptx.util import Inches

# PowerPointプレゼンテーションを作成する
prs = Presentation()

# スライドのレイアウトを選択する
title_slide_layout = prs.slide_layouts[0]
title_body_slide_layout = prs.slide_layouts[1]
body_slide_layout = prs.slide_layouts[6]
table_slide_layout = prs.slide_layouts[5]

# タイトルスライドを追加する
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
subtitle = slide.placeholders[1]
title.text = "企画書: 新商品の開発"
subtitle.text = "作成日: 2023年4月27日"

# スライド2を追加する (概要)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "概要"
body.text = "当社は、現在人気のある商品に代わる新商品の開発を計画しています。この新商品は、既存商品とは異なるデザインや機能を備え、新たな顧客層を獲得することを目指します。"

# スライド3を追加する (目的)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "目的"
body.text = "この企画の目的は、以下の通りです。\n\n- 既存商品に飽きた顧客層の獲得\n- 新商品の販売による売り上げ増加\n- 当社のブランド価値の向上"

# スライド4を追加する (計画)
slide = prs.slides.add_slide(title_slide_layout)
title = slide.shapes.title
title.text = "計画"

# スライド5を追加する (商品開発)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "商品開発"
body.text = "1. 新商品のコンセプト決定\n2. デザインの決定\n3. 機能の決定\n4. 原材料の調達"

# スライド6を追加する (試作品の開発)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "試作品の開発"
body.text = "1. 3Dプリンターを使用した試作品の製作\n2. 試作品のテスト\n3. 問題点の修正"

# スライド7を追加する (商品の製造)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "商品の製造"
body.text = "1. 生産ラインの設計\n2. 製造プロセスの決定\n3. 製造工場の設立"

# スライド8を追加する (販売)
slide = prs.slides.add_slide(title_body_slide_layout)
title = slide.shapes.title
body = slide.placeholders[1]
title.text = "販売"
body.text = "1. 販売戦略の決定\n2. 広告宣伝活動の実施\n3. 販売チャネルの確立"

# スライド9を追加する (予算)
slide = prs.slides.add_slide(table_slide_layout)
title = slide.shapes.title
title.text = "予算"
table = slide.shapes.add_table(rows=5, cols=2, left=Inches(0.5), top=Inches(1.5), width=Inches(8.0), height=Inches(1.0)).table
table.cell(0, 0).text = "項目"
table.cell(0, 1).text = "金額"
table.cell(1, 0).text = "商品開発費用"
table.cell(1, 1).text = "10,000,000円"
table.cell(2, 0).text = "試作品開発費用"
table.cell(2, 1).text = "5,000,000円"
table.cell(3, 0).text = "商品製造費用"
table.cell(3, 1).text = "50,000,000円"
table.cell(4, 0).text = "広告宣伝費用"
table.cell(4, 1).text = "20,000,000円"

# スライド10を追加する (スケジュール)
slide = prs.slides.add_slide(table_slide_layout)
title = slide.shapes.title
title.text = "スケジュール"
table = slide.shapes.add_table(rows=5, cols=2, left=Inches(0.5), top=Inches(1.5), width=Inches(8.0), height=Inches(1.0)).table
table.cell(0, 0).text = "項目"
table.cell(0, 1).text = "期間"
table.cell(1, 0).text = "商品開発"
table.cell(1, 1).text = "3ヶ月"
table.cell(2, 0).text = "試作品開発"
table.cell(2, 1).text = "1ヶ月"
table.cell(3, 0).text = "商品製造"
table.cell(3, 1).text = "6ヶ月"
table.cell(4, 0).text = "広告宣伝"
table.cell(4, 1).text = "2ヶ月"

# ファイルに保存する
prs.save("企画書.pptx")
```
